```python
import pandas as pd
```

```python
data = pd.read_excel('products_dispatched_wagon_wise.xlsx')
```

```python
data
```

```python
df = data[['Mat Code', 'Wagon No']]
```

```python
df
```

```python
n = 10

grouped_data = [df[i:i+n] for i in range(0, len(df), n)]
concatenated_df = pd.concat(grouped_data, axis=1)
```

```python
transposed_df = concatenated_df.T
```

```python
transposed_df
```

```python
transposed_df.columns = transposed_df.iloc[0]
```

```python
transposed_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
categorical_columns = ['Wagon No']  # Replace with your column names
for col in categorical_columns:
    transposed_df[col] = label_encoder.fit_transform(transposed_df[col])
```

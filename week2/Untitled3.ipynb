```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Read csv file
df = pd.read_csv('mat,wagonremoved.csv')

# Define columns to impute
cols_to_impute = ['Mat Code', 'Heat No'] 

# Fill missing values with 'Unknown'
df[cols_to_impute] = df[cols_to_impute].fillna('Unknown')

# Create imputer and fit 
imputer = SimpleImputer(missing_values='Unknown', strategy='most_frequent')
imputer.fit(df[cols_to_impute])

# Transform data by replacing 'Unknown' values
df[cols_to_impute] = imputer.transform(df[cols_to_impute])

# Convert arrays back to dataframe
df[cols_to_impute] = pd.DataFrame(df[cols_to_impute], columns=cols_to_impute)

# Print final dataframe 
print(df)
```

```python
import pandas as pd
from sklearn.impute import KNearestNeighborsImputer  
from sklearn.preprocessing import OrdinalEncoder

# Load data
df = pd.read_csv('data.csv')

# Categorical columns to encode  
categorical_cols = ['Mat Code', 'Wagon No']

# Create ordinal encoder
encoder = OrdinalEncoder()

# Encode categorical columns
df[categorical_cols] = encoder.fit_transform(df[categorical_cols])

# Define columns to impute
cols_to_impute = ['Mat Code', 'Heat No'] 

# Additional features to use for imputation  
add_feats = ['Wagon No']

# Combine all features
feats = cols_to_impute + add_feats  

# Create KNN Imputer
imputer = KNearestNeighborsImputer(n_neighbors=5)

# Fit on and transform the data
df[cols_to_impute] = imputer.fit_transform(df[feats])  

# Convert back to dataframe
df[cols_to_impute] = pd.DataFrame(df[cols_to_impute], columns=cols_to_impute)

# Print output
print(df)
```
